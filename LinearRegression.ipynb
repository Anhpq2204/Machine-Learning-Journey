Linear regression

```python
#import necessary library
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
#load the dataset
X,y= datasets.make_regression(n_samples=100,n_features=1,noise=20,random_state=4)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4)
X_train.shape
```

```python
y_train.shape
```

```python
# the model ( lr, n_iter, weight, bias)
#formula y^= w*x+b
class Linear_regression:
    def __init__(self,lr=0.001,n_iter=1000):
        self.lr=lr
        self.n_iter=1000
        self.wei=None
        self.bias=None
    def fit(self,X,y):
        self.X_train=X
        self.y_train=y
        n_sample,n_features=self.X_train.shape
        self.bias=0
        self.wei=np.zeros(n_features)
        #interation for gradient descent
        for i in range(0,self.n_iter):
            y_predicted=np.dot(self.X_train,self.wei)+self.bias
            #print(y_predicted.shape)
            dw=(1/n_sample)*2*np.dot(self.X_train.T,y_predicted-self.y_train)
            db=(1/n_sample)*2*np.sum(y_predicted-self.y_train)
            self.wei-=self.lr*dw
            self.bias-=self.lr*db
    def predict(self,X):
        return np.dot(X,self.wei)+self.bias

md=Linear_regression(lr=0.01)
md.fit(X_train,y_train)
predicted=md.predict(X_test)
def mse(y_real,y_predicted):
    return np.mean((y_real-y_predicted)**2)
print(mse(y_test,predicted))
        
        
        
```

```python
#visualize
import matplotlib.pyplot as plt
# Plotting Function
def plot_regression(X, y, model):
    plt.scatter(X, y, color="blue", label="Data Points")
    
    # Sorting X for line plot
    X_line = np.sort(X, axis=0)
    y_line = model.predict(X_line)
    
    plt.plot(X_line, y_line, color="red", label="Regression Line")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.title("Linear Regression Fit")
    plt.legend()
    plt.show()
plot_regression(X_train, y_train, md)
```
